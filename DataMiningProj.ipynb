# Pearson's Correlation

```python
def correlation(dataset, threshold):
    col_corr = []  
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (abs(corr_matrix.iloc[i, j]) > threshold):
              colname = corr_matrix.columns[i]                
              col_corr.append(colname)
    return col_corr
corr_features = correlation(df, 0.85)
corr_features
```

### K-MEANS CLUSTERING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# set number of clusters
K = 2

# initialize centroids
centroids = df.sample(K)
centroids.index = range(K)

# create scatterplot
colors = np.where(df['Lung problems'] == 1, 'blue','red')
markers = 'o'
plt.scatter(df['SB18S'], df['M19'], c=colors, marker=markers)

# run k-means algorithm
cluster_labels = np.zeros(len(df))
for iteration in range(10):
    # calculate distance matrix
    m = len(df)
    dist_matrix = np.zeros((m, K))
    for i in range(K):
        dist_matrix[:, i] = np.linalg.norm(df[['SB18S', 'M19']].values - centroids.iloc[i, [2, 3]].values.reshape(1, -1), axis=1)

    # assign clusters
    new_labels = np.argmin(dist_matrix, axis=1)
    if np.array_equal(new_labels, cluster_labels):
        break
    else:
        cluster_labels = new_labels

    # update centroids
    for i in range(K):
        centroids.iloc[i][['SB18S', 'M19']] = df[cluster_labels == i][['SB18S', 'M19']].mean().values.copy()

# assign clusters
cluster_labels = np.argmin(dist_matrix, axis=1).astype(int)
df['cluster'] = cluster_labels

# calculate cluster cardinality
cluster_sizes = np.bincount(cluster_labels)

# plot histogram of cluster cardinality
plt.hist(cluster_labels, bins=range(K+1))
plt.xticks(range(K))
plt.xlabel('Cluster')
plt.ylabel('Number of data points')
plt.show()

# create scatterplot with final clusters
colors = np.where(cluster_labels == 0, 'blue','red')
markers = 'x'
plt.scatter(df['SB18S'], df['M19'], c=colors, marker=markers)
plt.show()

# print final cluster magnitude
print('Cluster Magnitude:', cluster_sizes)
```
