```python
def correlation(dataset, threshold):
    col_corr = []  
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (abs(corr_matrix.iloc[i, j]) > threshold):
              colname = corr_matrix.columns[i]                
              col_corr.append(colname)
    return col_corr
corr_features = correlation(df, 0.85)
corr_features
```

```python
import numpy as np

class KMeans:
    def __init__(self, n_clusters, max_iter=300):
        self.n_clusters = n_clusters
        self.max_iter = max_iter

    def fit(self, df):
        # Step 1: Initialize centroids
        centroids = df.sample(n=self.n_clusters).values

        for i in range(self.max_iter):
            # Step 2: Assign data points to centroids
            distances = np.sqrt(((df.values - centroids[:, np.newaxis])**2).sum(axis=2))
            labels = np.argmin(distances, axis=0)

            # Step 3: Update centroids
            new_centroids = np.array([df.values[labels == j].mean(axis=0) for j in range(self.n_clusters)])

            # Check convergence
            if np.allclose(centroids, new_centroids):
                break

            centroids = new_centroids

        # Step 5: Return the centroids and labels as DataFrames
        centroids_df = pd.DataFrame(centroids, columns=df.columns)
        labels_df = pd.DataFrame({'label': labels})

        return centroids_df, labels_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# calculate distortion for a range of number of cluster values
distortions = []
K_range = range(1, 10)
for K in K_range:
    # initialize centroids
    centroids = df.sample(K)
    centroids.index = range(K)

    # run k-means algorithm
    for iteration in range(10):
        # calculate distance matrix
        m = len(df)
        dist_matrix = np.zeros((m, K))
        for i in range(K):
            dist_matrix[:, i] = np.linalg.norm(df[['SB18S', 'SB18D']].values - centroids.iloc[i, [2, 3]].values.reshape(1, -1), axis=1)

        # assign clusters
        cluster_labels = np.argmin(dist_matrix, axis=1)
        df['cluster'] = cluster_labels

        # update centroids
        for i in range(K):
            centroids.iloc[i][['SB18S', 'SB18D']] = df[df['cluster'] == i][['SB18S', 'SB18D']].mean().values.copy()

    # calculate distortion
    distortion = np.sum(np.min(dist_matrix, axis=1)) / m
    distortions.append(distortion)

# plot elbow curve
plt.plot(K_range, distortions, 'bx-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Distortion')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
import pandas as pd
import numpy as np

def create_C1(data):
    """
    Create a list of unique items in the dataset.
    """
    C1 = []
    for transaction in data:
        for item in transaction:
            if not [item] in C1:
                C1.append([item])
    C1.sort()
    return [frozenset([item]) for item in C1]

def support_prune(Ck, D, min_support):
    """
    Returns all candidates in Ck that have a support >= min_support.
    """
    # Count the support of each candidate
    item_count = {}
    for transaction in D:
        for item in Ck:
            if item.issubset(transaction):
                if item not in item_count:
                    item_count[item] = 1
                else:
                    item_count[item] += 1
    
    # Prune candidates that do not meet the minimum support
    frequent_items = []
    support_data = {}
    n_transactions = float(len(D))
    for item in item_count:
        support = item_count[item] / n_transactions
        if support >= min_support:
            frequent_items.append(item)
            support_data[item] = support
    return frequent_items, support_data

def generate_candidates(Lk, k):
    """
    Generate candidate itemsets of length k from frequent itemsets of length k-1.
    """
    # Initialize variables
    candidates = []
    n = len(Lk)
    
    # Join Lk with itself to create candidate sets
    for i in range(n):
        for j in range(i+1, n):
            l1 = list(Lk[i])[:k-2]
            l2 = list(Lk[j])[:k-2]
            l1.sort()
            l2.sort()
            if l1 == l2:
                candidates.append(Lk[i] | Lk[j])
    return candidates

def generate_frequent_itemsets(data, min_support):
    """
    Generate all frequent itemsets that have a support >= min_support.
    """
    # Initialize variables
    C1 = create_C1(data)
    D = [set(row) for row in data]
    L1, support_data = support_prune(C1, D, min_support)
    L = [L1]
    k = 2
    
    # Generate candidate itemsets and prune them until no more frequent itemsets can be found
    while len(L[k-2]) > 0:
        candidates = generate_candidates(L[k-2], k)
        ck, support_ck = support_prune(candidates, D, min_support)
        support_data.update(support_ck)
        L.append(ck)
        k += 1
    
    # Convert the support data to a pandas DataFrame for easy access
    support_df = pd.DataFrame.from_dict(support_data, orient='index', columns=['support'])
    
    # Return the frequent itemsets with their corresponding support data
    frequent_itemsets = []
    for itemset in L:
        for item in itemset:
            frequent_itemsets.append((item, support_data[item]))
    frequent_itemsets = sorted(frequent_itemsets, key=lambda x: x[1], reverse=True)
    
    return frequent_itemsets



frequent_itemsets = generate_frequent_itemsets(df.values, min_support=0.5)

for itemset in frequent_itemsets:
    print(itemset)
```
